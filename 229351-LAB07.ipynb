### ปฏิบัติการครั้งที่ 7
1. สร้างโมเดลเพื่อการทำนายดังนี้
1.1 แบ่งข้อมูลออกเป็น 2 ส่วน
   - training set: วันที่ 1996-01-01 ถึง 2009-12-01
   - test set: วันที่ 2010-01-01 ถึง 2012-03-01  
1.2 แยกส่วนประกอบ $y_t=S_t+T_t+R_t$ บน training set
1.3 สร้าง time series ชุดใหม่ที่แสดงถึงทำนายค่าบน test set โดยนำค่า $T_t+R_t$ ของวันล่าสุดใน training set ที่มีค่า $T_t$ มาบวกกับแต่ละค่าใน $S_t$ จากวันที่ วันที่ 2010-01-01 ถึง 2012-03-01  
3. คำนวณ RMSE โดยใช้ฟังก์ชัน `rmse` ข้างล่าง
4. แสดงแผนภาพข้อมูล elecequip และค่าทำนายที่ได้

```python
data.dropna()

#Training set
tr =data.loc['1996-01-01':'2009-12-01']
```

```python
#Test set
test = data.loc['2010-01-01' : '2012-03-01']
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result_add_tr = seasonal_decompose(tr['value'], model='additive')
s_tr = result_add_tr.seasonal
t_tr = result_add_tr.trend
r_tr = result_add_tr.resid

last_trend = t_tr.dropna().iloc[-1]


result_add_full = seasonal_decompose(data['value'], model='additive', period=12)
s_test = result_add_full.seasonal.loc[test.index]


y = s_test + last_trend
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))
```

```python
plt.figure(figsize=(18,6))
plt.plot(data.index,data['value'],label='Original Data')
plt.plot(test.index, y, marker='o')
plt.axvline(pd.to_datetime("2010-01-01"), color='red', label='Prediction Start')
plt.legend()
plt.title('Elecequip Data and Prediction')
plt.show()
```
